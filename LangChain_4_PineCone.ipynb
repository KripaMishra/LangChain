# Extracting Text from pdf files


```python
from langchain.document_loaders import PyPDFDirectoryLoader
```

```python
loader=PyPDFDirectoryLoader('pdfs')
```

```python
data=loader.load()
```

```python
data
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
text_splitter=RecursiveCharacterTextSplitter(chunk_size=400, chunk_overlap=10)
```

```python
text_chunk=text_splitter.split_documents(data)
```

```python
len(text_chunk)
```

```python
text_chunk[5]
```

# Embeddings

```python
from langchain.embeddings import HuggingFaceEmbeddings

embeddings=HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
query_result= embeddings.embed_query("jai bheem ")
```

```python
query_result
```

```python
len(query_result)
```

# Initializing the Pinecone

```python
from google.colab import userdata
```

```python
PINECONE_API_KEY = userdata.get("PINECONE_API_KEY")
PINECONE_API_ENV = userdata.get("PINECONE_API_ENV")
```

```python
import pinecone
# initializing the pinecone

pinecone.init(
    api_key=PINECONE_API_KEY,
    environment=PINECONE_API_ENV)

index_name="first"
```

```python
from langchain.vectorstores import Pinecone
```

```python
docsearch=Pinecone.from_texts([t.page_content for t in text_chunk],embeddings, index_name=index_name)
```

Similarity Search

```python
docsearch = Pinecone.from_existing_index(index_name, embeddings)
docsearch
```

```python
query = "skill"
```

```python
docs = docsearch.similarity_search(query, k=3)
```

```python
docs
```
